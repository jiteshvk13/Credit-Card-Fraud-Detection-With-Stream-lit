```python
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(X)
```

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score, accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print("Accuracy Score:",accuracy_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

# Calculate scale_pos_weight = (number of non-fraud) / (number of fraud)
scale_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

model = XGBClassifier(scale_pos_weight=scale_weight, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print("Accuracy Score:",accuracy_score(y_test, y_pred))
```

```python
from lightgbm import LGBMClassifier

model = LGBMClassifier(is_unbalance=True, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print("Accuracy Score:",accuracy_score(y_test, y_pred))
```

```python
model = LGBMClassifier(class_weight='balanced', random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
print("Accuracy Score:",accuracy_score(y_test, y_pred))
```

```python
y_proba = model.predict_proba(x_test)[:, 1]  # Probabilities for class 1 (fraud)
threshold = 0.52  # Try 0.3, 0.4, 0.6 etc.
y_pred_custom = (y_proba > threshold).astype(int)

print(classification_report(y_test, y_pred_custom))
```
